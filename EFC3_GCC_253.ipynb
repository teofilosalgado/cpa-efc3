### Questão 2

Uma parte importante para provar que SAT é redutível a 3SAT é tomar uma fórmula $C$ como instância de SAT e construir uma fórmula $C'$ para instância de 3SAT, tal que $C$ é satisfatível se e somente se $C'$ é satisfatível.

Nessa construção de $C'$ a partir de $C$, para cada cláusula $c_{j}=z_{1} \vee z_{2}  \vee ... \vee z_{m}$ de $C$ com $ m \gt 3$ literais faz-se uma coleção de $m - 2$ cláusulas com 3 literais cada para compor $C'$, introduzindo-se $U'_{j} = \{ y^{i}_{j}: 1 \leq i \leq m -3 \}$ com $m - 3$ variáveis novas.

Consideramos os seguintes problemas parametrizados SAT e 3SAT, os quais tomamos parâmetros nos problemas computacionais SAT e 3SAT do seguinte modo:

**SAT**

Instância: Uma fórmula $F$ em $CNF$ e um inteiro positivo $k$.

Parâmetro: O inteiro positivo $k$.

Pergunta: Há valoração de peso $k$ que satisfaz $F$?

**3SAT**

Instância: Uma fórmula $F$ em $3CNF$ e um inteiro positivo $k$.

Parâmetro: O inteiro positivo $k$.

Pergunta: Há valoração de peso $k$ que satisfaz $F$?

Usando a redução de SAT $\alpha$ 3SAT para tentar provar se SAT $\leq_{rp}$ 3SAT, percebe-se que os parâmetros não são redutíveis entre si. Suponha que a fórmula $C$ da demonstração de que SAT $\alpha$ 3SAT tenha uma valoração de peso $k$ que a satisfaz, tal que exatamento uma variável $u_{p}$ na cláusula $C_{j}$ seja verdadeira. Daí, para $C'$ ser satisfatível é necessário fazer com que $y^{1}_{j}...y^{p-2}_{j}$ seja todas verdadeiras. Portanto o peso da valoração que satisfaz $C'$ não depende apenas do parâmetro $k$ da valoração de $C$, mas depende de $p$ e daí depende do tamanho das cláusulas de $C$.

Como vimos na definição do conceito de redutíbilidade, o parâmetro 3SAT deve depender apenas do parâmetro de SAT para que SAT $\leq_{rp}$ 3SAT. Disto e dos argumentos acima não podemos usar a redução de SAT $\alpha$ 3SAT para fazer com que SAT $\leq_{rp}$ 3SAT.

Não é conhecido se SAT $\leq_{rp}$ 3SAT por alguma redução parametrizada e conjectura-se que não existe uma tal redução parametrizada.

### Questão 3

Podemos definir o problema 3SAT com a seguinte pergunta: Dada uma coleção _C_ de cláusulas booleanas, cada
uma com no máximo 3 literais, existe uma valoração que satisfaz todas as
cláusulas em _C_? Ou seja, supondo que _C_ seja definido por $(x_{1} \vee x_{2} \vee x_{3}) \wedge (x_{4}\vee x_{5} \vee x_{6})$, a questão a ser respondida é se existem valores de $x_{1}...x_{6}$ que resultariam num valor verdadeiro ao processamento das expressões booleanas. 

Antes de definirmos o problema CLIQUE precisamos definir o que é um clique. Denomina-se clique de um grafo G a um subgrafo (induzido) de G que seja completo. Dito isso, o problema refere-se a qualquer problema que possui como objetivo encontrar subgrafos completos (cliques) em um grafo. Como exemplo, o problema de encontrar conjuntos de nós em que todos os elementos estão conectados entre si.

Para provar que um problema é NP-Completo, temos que mostrar que ele pertence às classes NP e NP-Difícil. (Uma vez que os problemas NP-Completos são problemas NP-Difícil que também pertencem ao NP). A seguir, desenvolveremos ambas as provas.

Se um problema pertence à classe NP, então deve ter verificabilidade em tempo polinomial, que é dado um certificado, devemos ser capazes de verificar em tempo polinomial se é uma solução para o problema.

**Prova:**

1. Certificado - Seja o certificado um conjunto $S$ consistindo de nós no clique e $S$ é um subgráfico de $G$.
2. Verificação - Temos que analisar se existe um clique de tamanho k no grafo. Portanto, verificar se o número de nós em $S$ é igual a $k$ leva um tempo $O(1)$. Verificar se cada vértice tem um grau de saída de $(k-1)$ leva um tempo $O(k^{2})$. (Como em um grafo completo, cada vértice está conectado a todos os outros vértices por meio de uma aresta). Portanto, o número total de arestas em um grafo completo equivale a ${}^kC_{2} = k \times \frac{k-1}{2})$. Ademais, para verificar se o grafo formado pelos $k$ nós em $S$ está completo ou não, leva-se um tempo $O (k^{2} ) = O(n^{2})$ (pois $k <= n$, onde $n$ é o número de vértices em $G$).

Um problema $L$ pertence a NP-Hard se todo problema NP é redutível a $L$ em tempo polinomial. Dito isso, demonstraremos que o problema de Satisfabilidade Booleana (SAT) se reduz ao problema clique.

**Prova:**

Seja a expressão booleana $F = (x_{1} \vee x_{2}) \wedge (x_{1}' \vee x_{2}') \wedge (x_{1} \vee x_{3})$ onde $x_{1}$, x 2 , x 3 são as variáveis e $x'$ denota o complemento de $x$. Deixe a expressão dentro de cada parênteses ser uma cláusula. Portanto, temos três cláusulas $C1$, $C2$ e $C3$. Considere os vértices como: $<x_{1}, 1>$; $<x_{2}, 1>$; $<x_{1}', 2>$; $<x_{2}', 2>$; $<x_{1}, 3>$; $<x_{3}, 3>$ onde o segundo termo em cada vértice denota o número da cláusula a que pertencem. Conectamos esses vértices de forma que:

1. Não há dois vértices pertencentes à mesma cláusula conectados.
2. Nenhuma variável está conectada ao seu complemento.

![grafo](q3.png)

Assim, o grafo $G(V, E)$ é construído de tal forma que $V = \{<a, i> | a \in C_{i} \}$ e $E = \{(<a, i>, <b, j>) | i \neq j; b \neq a'\}$, considere o subgrafo de $G$ com os vértices $<x_{2}, 1>$; $<x_{1}', 2>$; $<x_{3} , 3>$. Ele forma um clique de tamanho 3 (representado pela linha pontilhada na figura acima). Correspondendo a isso, para a atribuição: $<x_{1}, x_{2}, x_{3}> = <0, 1, 1>$, $F$ resulta como verdadeiro. Portanto, se temos $k$ cláusulas em nossa expressão de satisfatibilidade, obtemos um clique máximo de tamanho $k$ e para a atribuição de valores correspondente, a expressão de satisfatibilidade é avaliada como verdadeira. Conseqeentemente, para uma instância particular, o problema de satisfatibilidade é reduzido ao problema de decisão de clique.

Portanto, o problema de decisão da Clique é NP-Difícil.

**Conclusão:**
O problema clique é NP e NP-Difícil. Portanto, o problema é NP-Completo.
   

### Questão 4

Problema da Mochila (em inglês, Knapsack Problem), utilizando programação dinâmica.

```python
import time

# Função para realizar o benchmark de uma função
def benchmark(function):
    def wrapper(*args, **kwargs):
        inicio = time.perf_counter()
        resultado = function(*args, **kwargs)
        fim = time.perf_counter()
        tempo_execucao = fim - inicio
        return resultado, tempo_execucao
    return wrapper

# Função para executar o problema
@benchmark
def knapsack(peso_maximo, pesos, valores):
    m = [[0 for x in range(peso_maximo + 1)] for x in range(len(valores) + 1)]

    for i in range(len(valores) + 1):
        for j in range(peso_maximo + 1):
            if i == 0 or j == 0:
                m[i][j] = 0
            elif pesos[i - 1] <= j:
                m[i][j] = max(valores[i - 1] + m[i - 1][j - pesos[i - 1]], m[i - 1][j])
            else:
                m[i][j] = m[i - 1][j]
    return m[len(valores)][peso_maximo]


# Driver code
valores = [60, 100, 120]
pesos = [10, 20, 30]
peso_maximo = 50
print(knapsack(peso_maximo, pesos, valores))
```
